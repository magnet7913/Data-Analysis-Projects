```python
import pandas as pd
```

```python
df = pd.read_csv('CovidDeaths.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
t1 = df[df['location']!='World'].groupby(by=['continent','location'])[['total_cases','total_deaths']].agg(max)
```

```python
totalCase = (t1.agg({'total_cases': 'sum', 'total_deaths': 'sum'}))
totalCase['mortality_ratio'] = totalCase['total_deaths'] / totalCase['total_cases']
```

```python
# Get total_cases, total_deaths and mortality_rates by country
totalCase.to_csv('mortality_rate.csv', index=True)
```

```python
# Get total Death by continent
t1.groupby(by='continent').agg({'total_deaths':'sum'}).to_csv('death_by_continent.csv',index=True)
```

```python
t2 = df[df['location']!='World'].groupby(by=['location'])[['population','total_cases']].agg(max)
```

```python
t2[t2['total_cases'].isna()]
```

```python
# We have got some country with no cases value, lets drop them
t2= t2[t2['total_cases'].notna()]
```

```python
# Calculate the % population infected
t2['PercentPopulationInfected'] = t2['total_cases']/t2['population']
```

```python
t2
```

```python
t2.to_csv('PopulationInfectedByCountry.csv', index=True)
```

```python
t3 = df[df['continent'].notna()].groupby(by=['location','date','population'])[['total_cases']].agg(max).reset_index()
```

```python
t3= t3[t3['total_cases'].notna()]
```

```python
t3['PercentPopulationInfected'] = t3['total_cases'] / t3['population']
```

```python
t3.to_csv('daily_cases_by_country.csv',index=True)
```
