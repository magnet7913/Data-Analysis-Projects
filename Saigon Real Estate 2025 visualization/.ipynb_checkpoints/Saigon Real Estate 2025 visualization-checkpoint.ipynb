```python
import time
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
import json
```

```python
geolocator = Nominatim(
    user_agent="vietnam_geocoder_getadid578@icousd.com",  # MUST be unique & include contact
    timeout=10
)

geocode = RateLimiter(
    geolocator.geocode,
    min_delay_seconds=1.1,      # Slightly over 1s to be safe
    max_retries=3,
    error_wait_seconds=5
)

cache_file = "nominatim_cache.json"
try:
    with open(cache_file, "r", encoding="utf-8") as f:
        cache = json.load(f)
except FileNotFoundError:
    cache = {}

results = []

for addr in dia_chi:
    try:
        location = geocode(
            query=addr,
            country_codes=["vn"],  # Restrict to Vietnam for accuracy
            addressdetails=True    # Optional: gets structured details
        )
        
        if location:
            lat = location.latitude
            lon = location.longitude
            print(f"Success: {addr} -> ({lat}, {lon})")
        else:
            lat = lon = None
            print(f"Not found: {addr}")
            
        cache[addr] = (lat, lon)
        
    except Exception as e:
        print(f"Error for {addr}: {e}")
        lat = lon = None
        cache[addr] = (None, None)
    
    results.append({"address": addr, "latitude": lat, "longitude": lon})
    
    if len(results) % 100 == 0:
        with open(cache_file, "w", encoding="utf-8") as f:
            json.dump(cache, f, ensure_ascii=False, indent=2)

with open(cache_file, "w", encoding="utf-8") as f:
    json.dump(cache, f, ensure_ascii=False, indent=2)

final_results = []
for addr in dia_chi:
    lat, lon = cache.get(addr, (None, None))
    final_results.append({"address": addr, "latitude": lat, "longitude": lon})

import pandas as pd
df = pd.DataFrame(final_results)
df.to_csv("geocoded_vietnam_addresses.csv", index=False, encoding="utf-8")

print("Geocoding complete. Results saved to CSV.")
```
